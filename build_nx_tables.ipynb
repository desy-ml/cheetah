```python
from pathlib import Path

import matplotlib.pyplot as plt

import cheetah
from cheetah.converters.nxtables import read_nx_tables
from tests.resources import ARESlatticeStage3v1_9 as ocelot_lattice
```

```python
nx_tables_path = Path("tests/resources/Stage4v3_9.txt")
nx_table_segment = read_nx_tables(nx_tables_path)
nx_table_segment.length
```

```python
beam = cheetah.ParameterBeam.from_astra("benchmark/astra/ACHIP_EA1_2021.1351.001")
```

```python
ocelot_segment = cheetah.Segment.from_ocelot(ocelot_lattice.cell)
```

```python
ocelot_segment.plot_twiss_over_lattice(beam)
nx_table_segment.plot_twiss_over_lattice(beam)
```

```python
ocelot_segment.subcell("AREASOLA1", "AREABSCR1").plot_twiss_over_lattice(beam)
nx_table_segment.subcell("AREASOLA1", "AREABSCM1").plot_twiss_over_lattice(beam)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 3), sharex=True)

ocelot_segment.subcell("AREASOLA1", "AREABSCR1").plot(axs[0], s=0.0)
nx_table_segment.subcell("AREASOLA1", "AREABSCM1").plot(axs[1], s=0.0)

axs[0].set_title("Experimental Area - Stage 3 v1.9")
axs[1].set_title("Experimental Area - Stage 4 v3.9")

axs[0].set_xlabel(None)
plt.tight_layout()

plt.show()
```
